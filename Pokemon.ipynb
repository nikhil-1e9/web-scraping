```python
import requests
from bs4 import BeautifulSoup
```

```python
BASE_URL = "https://pokemondb.net"
```

```python
resp = requests.get(url="https://pokemondb.net/pokedex/all")
resp.status_code
```

```python
doc = BeautifulSoup(resp.text, 'html.parser')
```

```python
poke_links = [x['href'] for x in doc.find_all(class_='ent-name')]
poke_links[:10]
```

```python
resp = requests.get(url=BASE_URL+poke_links[2])
poke_doc = BeautifulSoup(resp.text, 'html.parser')
```

```python
poke_doc.find_all(class_='vitals-table')[0].find_all('td')[0].text
```
